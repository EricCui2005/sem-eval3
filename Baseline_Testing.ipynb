```python
#@title Imports

import torch
from transformers import GPT2Tokenizer, GPT2ForSequenceClassification
import transformers
import json
```

```python
#@title load_emotion_classifier()

def load_emotion_classifier():

    # Initializing the tokenizer that will extract the tokens from the sequences
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

    # Initializing the classification model
    model = GPT2ForSequenceClassification.from_pretrained('gpt2', num_labels=6)
    return model, tokenizer
```

```python
#@title classify_emotion()

def classify_emotion(model, tokenizer, text):

    # Dictionary containing various key-value pairs
    # "id_mask" assigns a unique integer identifier to each token
    # "attention_mask" is a tensor that is the same length as the tokenized tensor
    # Contains 1 for tokens that should be attended to and 0 for tokens that should be ignored
    inputs = tokenizer(text, return_tensors="pt")
    # print(inputs)

    # I'm scared of the raw output
    outputs = model(**inputs)

    # Raw predicted scores for each emotion
    logits = outputs.logits
    print(logits)

    # Finds the index of the maximum value along the second dimension of the tensor
    # .item() converts it to an integer
    predicted_emotion = torch.argmax(logits, dim=1).item()

    return predicted_emotion
```

```python
#@title identify_cause_span()

def identify_cause_spans(text, predicted_emotion):
    if predicted_emotion == 0:  #"Anger"
        return [text]
    elif predicted_emotion == 1:  #"Disgust"
        return [text]
    elif predicted_emotion == 2:  #"Fear"
        return [text]
    elif predicted_emotion == 3:  #"Joy"
        return [text]
    elif predicted_emotion == 4:  #"Sadness"
        return [text]
    elif predicted_emotion == 5:  #"Surprise"
        return [text]
    else:
        return []
```

```python
#@title extract_emotion_cause_pairs

def extract_emotion_cause_pairs(conversation_data):

    # Initializing the model, tokenizer, and array to store emotion-cause pairs
    emotion_model, tokenizer = load_emotion_classifier()
    emotion_cause_pairs = []

    # Looping through all utterances in the dataset
    for utterance in conversation_data["utterances"]:

        # Emotion prediction and cause span extraction
        predicted_emotion = classify_emotion(emotion_model, tokenizer, utterance["text"])
        cause_spans = identify_cause_spans(utterance["text"], predicted_emotion)

        print("Utterance:", utterance["text"])
        print("Predicted Emotion:", predicted_emotion)
        print("Cause Spans:", cause_spans)

        for cause_span in cause_spans:
            emotion_cause_pairs.append((f"{utterance['utterance_ID']}_{predicted_emotion}", f"{utterance['utterance_ID']}_{cause_span}"))

    return emotion_cause_pairs
```

```python
#@title Preprocessing Snippet Training Data

# Array to store text/tagged data
training_snippet_processed = []

# Iterating through JSON
for conversation in training_snippet:
  for transcript in conversation["conversation"]:
    training_snippet_processed.append(((transcript["text"]), (transcript["emotion"])))
print(training_snippet_processed)
```

```python
#@title The Bigger Boy

from transformers import GPT2Tokenizer, GPT2ForSequenceClassification, AdamW
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import torch

# Initializing tokenizer and classification model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2ForSequenceClassification.from_pretrained('openai/whisper-large-v3', num_labels=7)
model.config.pad_token_id = model.config.eos_token_id

# Padding to ensure the tokeznied sequences are the same
tokenizer.pad_token = tokenizer.eos_token

# Processing the data
dataset = training_snippet_processed  # Assuming 'text_emotion' is your dataset

# Tuples
texts, labels = zip(*dataset)

# encoded_texts is an instance of the 'BatchEncoding' class from the Hugging Face 'transformers' library
encoded_texts = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')

# 2-D Tensor (Matrix)
# Extracting the tokenized text sequences
input_id_text = encoded_texts['input_ids']

# Converting the emotions in the label tuple to numbers and converting the list into a tensor
# 1-D tensor (Vector)
labels = torch.tensor([{'fear': 0, 'sadness': 1, 'surprise': 2, 'joy': 3, 'disgust': 4, 'neutral': 5, 'anger': 6}[label] for label in labels])

# Function splits the input_id_text and corresponding labels into two "pairs"
# One train/val pair is used for training, and the other pair is used for model evaluation
# test_size = 0.2 indicates that 80% of the model is used for training while the remaining 20% is reserved for testing
# Sets are all 2-D Tensors
train_texts, val_texts, train_labels, val_labels = train_test_split(input_id_text, labels, test_size=0.2)

# TensorDataset object
# Each element is a tuple containing individual tensors
# In this case, it is train_texts and train_labels
train_dataset = TensorDataset(train_texts, train_labels)

# Initializing the optimizer
# model.parameter() returns an iterator over all the model parameters
optimizer = AdamW(model.parameters(), lr=1e-5)

# Change this to something bigger than 1
batch_size = 32
epochs = 5

# DataLoader object
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Iterating for the specified number of epochs
for epoch in range(epochs):

     # Setting the model to training mode
     model.train()

     # Iterating over batches of data
     for batch in train_loader:

         # Extracting the Tensors containing the input_ids and
         # corresponding emotion labels
         input_ids, labels = batch

         # Resetting the gradients of all model parameters to calculate new gradients
         optimizer.zero_grad()

         # Passing in the input sequences ('input_ids') through the model
         # The model computs the predicted outputs based on the input data
         # 'labels' are provided as ground truth training
         outputs = model(input_ids, labels=labels)

         # Retrieving the los from the model's output
         loss = outputs.loss

         # Backward pass
         # Computing the gradients of the model parameters with respect to the loss
         # Gradients are then used to update the model weights in the direction that
         # minimizes loss
         loss.backward()

         # Updating the model parameters using the optimizer
         optimizer.step()
```
